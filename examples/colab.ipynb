# Example plotting

````python
from plotProfile import ReactionProfilePlotter

energy_sets = {
    "Pathway A": [0.00, -2.0, 10.2, 1.4, -1.5, 2.0, -7.2],
    "Pathway B": [None, -2.0, 6.2, 4.3, 5.8, 2.0],
}

annotations = {
    'Step 1': (0,3),
    'Step 2': (3,5),
    'Step 3': (5,6),
}

plotter = ReactionProfilePlotter(dashed=["Pathway B"])
plotter.plot(energy_sets, annotations=annotations)
````

````python
from plotProfile import ReactionProfilePlotter

energy_sets = {
    "Pathway A": [0.00, -2.0, 10.2, 1.4, -1.5, 2.0, -7.2],
    "Pathway B": [None, -2.0, 6.2,None, 3.8, 2.0],
    "diastereomer": [ None, None, 12.0],
    "Pathway C": [None, None, None, 1.4, 3.0, 3.0]
}

annotations = {
    'Step 1': (0,3),
    'Step 2': (3,5),
    'Step 3': (5,6),
}

plotter = ReactionProfilePlotter(dashed=["Pathway B"], point_type='bar', colors='viridis', curviness=0.5, axes=None )
plotter.plot(energy_sets, annotations=annotations, exclude_from_legend=["Pathway B"])
````
