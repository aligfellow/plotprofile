```python
import sys
from pathlib import Path
sys.path.append(str(Path("..") / "src"))

import plotProfile
import numpy as np

energy_sets = {
    "Pathway A": [0.00, -2.0, 4.1, 5.2, -8.3, -7.6, -5.0,-9.4,-4.2],
    "Pathway B": [0.00, 9.9, 9.9, 9.9, 9.9, -7.6, -2.5,-5,-4.2],
    "off-cycle": [np.nan, -2.0, -7.0, -10.4],
    "demo": [np.nan, np.nan, np.nan, -10.4, -13, -11, -11],
    "branching": [np.nan, np.nan, np.nan, np.nan, -8.3, -11, -11.0, -12,-12],
}

annotations = {
    'Step 1': (0,3),
    'Step 2': (3,6),
    'Step 3': (6,8),
}

plotter = plotProfile.plot.ReactionProfilePlotter(style="default", dashed=["off-cycle", "branching"], point_type='hollow', desaturate=True, segment_annotations=annotations, axes="box")
plotter.plot(energy_sets)

plotter = plotProfile.plot.ReactionProfilePlotter(style="presentation", dashed=["off-cycle", "branching"], point_type='dot', desaturate=False, segment_annotations=annotations, axes="box", colors='Greens_r', annotation_color='black')
plotter.plot(energy_sets)

plotter = plotProfile.plot.ReactionProfilePlotter(figsize=(6,2), dashed=["off-cycle", "branching"], point_type='bar', desaturate=True, segment_annotations=annotations, axes=None, colors=['red', 'blue', 'green', 'orange', 'purple'])
plotter.plot(energy_sets)
```
